```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
##Data cleaning
##missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
df=dataset
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype('int')
```

```python
df.head
```

```python
##Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##remove the row with index 122
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
### change data type
df[['month','day','Temperature','RH','Ws','year']]=df[['month','day','Temperature','RH','Ws','year']].astype('int') 

```

```python
df.info()
```

```python
## changing others columns to float
df[['FFMC','DMC','DC','ISI','BUI','FWI']]=df[['FFMC','DMC','DC','ISI','BUI','FWI']].astype('float')
```

```python
df[['Rain']]=df[['Rain']].astype('float')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```

```python
## plot density plot for all features
print(plt.style.available)
plt.style.use('seaborn-v0_8')
plt.style.use('seaborn-v0_8-darkgrid')  # Or try 'seaborn-v0_8-whitegrid'

# Plot histogram for all features in df_copy
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
##percentaage of pie chart
percentage = df_copy['Classes'].value_counts(normalize=True) * 100
# Plot pie chart  
plt.figure(figsize=(8, 6))
plt.pie(percentage, labels=percentage.index, autopct='%1.1f%%', startangle=140)

plt.title('Percentage of Classes')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```
