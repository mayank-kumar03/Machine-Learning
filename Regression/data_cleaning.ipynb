```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
##Data cleaning
##missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
df=dataset
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df.info()
```

```python
##Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
### change data type
df[['month','day','Temperature','RH','Ws','year']]=df[['month','day','Temperature','RH','Ws','year']].astype('int') 

```

```python
df.info()
```

```python
## changing others columns to float
df[['FFMC','DMC','DC','ISI','BUI','FWI']]=df[['FFMC','DMC','DC','ISI','BUI','FWI']].astype('float')
```

```python
df[['Rain']]=df[['Rain']].astype('float')
```
