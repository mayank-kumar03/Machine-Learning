```python
corpus=""" Hello welcome ,to mayank NLP practice.
This is a simple text corpus's for testing purposes.
It contains various sentences, punctuation, and some numbers like 123 and 456.
"""
```

```python
print(corpus)
```

```python
import nltk

nltk.download('punkt_tab')
```

```python
nltk.data.path.append('/D')
```

```python
##Tokenization
##sentence -->paragraph
from nltk.tokenize import sent_tokenize, word_tokenize
sentences = sent_tokenize(corpus)
print(sentences)
```

```python
type(sentences)
```

```python
##sentences->#words
words = [word_tokenize(sentence) for sentence in sentences]
print(words)
```

```python
from nltk.tokenize import wordpunct_tokenize
word_punct = [wordpunct_tokenize(sentence) for sentence in sentences]
print(word_punct)
```

```python
from nltk.tokenize import TreebankWordTokenizer 
detokenizer = TreebankWordTokenizer()
detokenized_words = [detokenizer.tokenize(sentence) for sentence in sentences]
print(detokenized_words)
```
