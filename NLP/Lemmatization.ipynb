```python
##Lemmatization 
#Lemmatization is the process of converting a word to its base or root form.
# For example, the lemmatization of "running" is "run", and the lemmatization of "better" is "good".
# This is different from stemming, which simply removes suffixes and prefixes to get to a base form, which may not be a valid word.
```

```python
#Wordnet Lemmatization
from nltk.stem import WordNetLemmatizer 
lematizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
print(lematizer.lemmatize("running")) # returns 'running' because it is not a valid word in WordNet
print(lematizer.lemmatize("running", pos="v")) # returns 'run' because it is a verb
print(lematizer.lemmatize("better", pos="a")) # returns 'good' because it is an adjective
print(lematizer.lemmatize("better", pos="n")) # returns 'better' because it is a noun
print(lematizer.lemmatize("better", pos="r")) # returns 'better' because it is an adverb
print(lematizer.lemmatize("better")) # returns 'better' because it is a noun
```
