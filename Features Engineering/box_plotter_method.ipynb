```python
##box_plotter-> it is a script to plot boxplots of the data and save them
import numpy as np
```

```python
lst_marks=[45,32,56,75,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.5,0.75,1])
print("Minimum:",minimum)
print("Q1:",Q1)
print("Median:",median)
print("Q3:",Q3)
print("Maximum:",maximum)
```

```python
iqr=Q3-Q1
print("IQR:",iqr)
```

```python
lower_bound=Q1-1.5*iqr
upper_bound=Q3+1.5*iqr
print("Lower Bound:",lower_bound)
print("Upper Bound:",upper_bound)
```

```python
import seaborn as sns
sns.boxplot(data=lst_marks)
```
