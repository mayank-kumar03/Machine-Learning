```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
iris_data=pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_iris = scaler.fit_transform(iris_data)
```

```python
scaled_iris
```

```python
##apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_iris = pca.fit_transform(scaled_iris)
```

```python
pca_iris
```

```python
plt.scatter(pca_iris[:, 0], pca_iris[:, 1], c=iris.target, cmap='viridis')
```

```python
##apply agglomerative clustering
##plot dendrogram
import scipy.cluster.hierarchy  as sc
#plot dendogram
plt.figure(figsize=(20, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
sc.dendrogram(sc.linkage(pca_iris, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
agglo_cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
agglo_cluster.fit(pca_iris)
```

```python
 agglo_cluster.labels_
# Plot the clusters
plt.scatter(pca_iris[:, 0], pca_iris[:, 1], c=agglo_cluster.labels_)
```
