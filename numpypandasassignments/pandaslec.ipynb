```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
df.head(5)
```

```python
df.dtypes
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_filled=df.fillna(0)
df_filled.isnull().any()
```

```python
#filling missing value
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
```

```python
df
```

```python
df.dtypes
```

```python
#renaming column
df=df.rename(columns={'Date':'Sale Date'})
df.head()
```

```python
#change data type
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['New_value']=df['Value'].apply(lambda x:x*2)
print(df)
```

```python
#data aggregating group
df.head()
grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
#merging  and joining data set 
df1=pd.DataFrame({'key':['A','B','C'],'Value1':[1,2,4]})
df2=pd.DataFrame({'key':['A','B','D'],'Value2':[1,2,4]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
